```python
# ============================================================
# 11. FEATURE ENGINEERING & PREPROCESSING
# ============================================================

"""
Preprocessing is one of the most critical stages in fraud detection
because improper preprocessing may lead to:

- data leakage
- misleading model performance
- biased fraud prediction
- deployment inconsistencies

This section focuses on:
- feature-target separation
- scaling strategy
- train-test splitting
- leakage prevention
- preprocessing consistency

Important Note:

SMOTE and imbalance handling will NOT be applied at this stage.
They will only be applied after train-test splitting to prevent
data leakage.
"""
```

```python
# ============================================================
# FEATURE & TARGET SEPARATION
# ============================================================

X = df.drop(
    
    'Class',
    
    axis=1
)

y = df['Class']

print("Feature Matrix Shape :", X.shape)

print("Target Vector Shape :", y.shape)
```

```python
# ============================================================
# CHECKING TARGET DISTRIBUTION
# ============================================================

print("Target Distribution :\n")

print(y.value_counts())

print("\n--------------------------------------------------\n")

print("Target Distribution Percentage :\n")

print(
    
    y.value_counts(normalize=True) * 100
)
```

```python
# ============================================================
# SCALING IMPORTANT FEATURES
# ============================================================

"""
Although PCA-transformed features are already scaled,
the following features still require preprocessing:

- Time
- Amount

These features have different scales and distributions
compared to PCA-transformed variables.

Scaling improves:
- model stability
- optimization performance
- distance-based learning
- anomaly detection sensitivity
"""
```

```python
# ============================================================
# IMPORTING SCALER
# ============================================================

from sklearn.preprocessing import StandardScaler
```

```python
# ============================================================
# COPYING DATA BEFORE TRANSFORMATION
# ============================================================

X_scaled = X.copy()
```

```python
# ============================================================
# APPLYING STANDARD SCALING
# ============================================================

scaler = StandardScaler()

X_scaled[['Amount', 'Time']] = scaler.fit_transform(
    
    X_scaled[['Amount', 'Time']]
)

print("Scaling Applied Successfully")
```

```python
# ============================================================
# VERIFYING SCALED FEATURES
# ============================================================

X_scaled[['Amount', 'Time']].describe().T
```

```python
# ============================================================
# OBSERVATIONS ON SCALING
# ============================================================

"""
Key Observations:

- Time and Amount features are now standardized
- Mean values are approximately centered around zero
- Scaling reduces magnitude differences between features

Business & Technical Importance:

- Prevents large-value dominance during training
- Improves optimization efficiency
- Enhances anomaly detection capability
- Helps distance-based algorithms perform better
- Improves deployment consistency
"""
```

```python
# ============================================================
# TRAIN-TEST SPLIT
# ============================================================

"""
Train-test splitting is performed before:
- SMOTE
- resampling
- anomaly modeling

This prevents:
- data leakage
- information contamination
- artificially inflated performance

Stratified sampling is used because:
- the dataset is highly imbalanced
- fraud class representation must be preserved
"""
```

```python
# ============================================================
# SPLITTING DATASET
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    
    X_scaled,
    
    y,
    
    test_size=0.2,
    
    random_state=42,
    
    stratify=y
)

print("Training Feature Shape :", X_train.shape)

print("Testing Feature Shape :", X_test.shape)

print("\n--------------------------------------------------\n")

print("Training Target Shape :", y_train.shape)

print("Testing Target Shape :", y_test.shape)
```

```python
# ============================================================
# VERIFYING STRATIFIED DISTRIBUTION
# ============================================================

print("Training Set Fraud Percentage :")

print(
    
    y_train.value_counts(normalize=True) * 100
)

print("\n--------------------------------------------------\n")

print("Testing Set Fraud Percentage :")

print(
    
    y_test.value_counts(normalize=True) * 100
)
```

```python
# ============================================================
# PREPROCESSING OBSERVATIONS
# ============================================================

"""
Major Preprocessing Outcomes:

1. Features and target variable successfully separated
2. Time and Amount features standardized
3. Leakage-safe preprocessing workflow maintained
4. Stratified train-test splitting successfully performed
5. Fraud class distribution preserved across train and test sets

Important Technical Note:

At this stage:
- NO SMOTE has been applied
- NO resampling has been performed

This is intentional because:
- resampling before splitting causes data leakage
- leakage produces unrealistic model performance

The current workflow follows industry-standard fraud
detection preprocessing practices.
"""
```

```python
# ============================================================
# 12. BASELINE MACHINE LEARNING MODELS
# ============================================================

"""
Baseline models are trained before applying SMOTE or imbalance handling.

Purpose of this stage:
- understand natural model behavior
- evaluate imbalance impact
- identify fraud detection limitations
- establish performance benchmark

This stage is extremely important because:
- fraud datasets are highly imbalanced
- models may ignore minority fraud class
- recall performance must be analyzed carefully

The following models will be evaluated:
1. Logistic Regression
2. Random Forest
3. Gradient Boosting
"""
```

```python
# ============================================================
# IMPORTING MACHINE LEARNING MODELS
# ============================================================

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import (
    
    RandomForestClassifier,
    
    GradientBoostingClassifier
)
```

```python
# ============================================================
# IMPORTING EVALUATION METRICS
# ============================================================

from sklearn.metrics import (
    
    accuracy_score,
    
    precision_score,
    
    recall_score,
    
    f1_score,
    
    roc_auc_score,
    
    classification_report,
    
    confusion_matrix,
    
    ConfusionMatrixDisplay
)
```

```python
# ============================================================
# CREATING EMPTY RESULTS LIST
# ============================================================

model_results = []
```

```python
# ============================================================
# MODEL EVALUATION FUNCTION
# ============================================================

def evaluate_model(model, model_name):
    
    # Training Model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    
    # Probabilities
    y_prob = model.predict_proba(X_test)[:,1]
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    
    precision = precision_score(y_test, y_pred)
    
    recall = recall_score(y_test, y_pred)
    
    f1 = f1_score(y_test, y_pred)
    
    roc_auc = roc_auc_score(y_test, y_prob)
    
    # Train Score
    train_score = model.score(X_train, y_train)
    
    # Test Score
    test_score = model.score(X_test, y_test)
    
    # Appending Results
    model_results.append({
        
        'Model' : model_name,
        
        'Accuracy' : accuracy,
        
        'Precision' : precision,
        
        'Recall' : recall,
        
        'F1 Score' : f1,
        
        'ROC-AUC' : roc_auc,
        
        'Train Score' : train_score,
        
        'Test Score' : test_score
    })
    
    # Printing Metrics
    print(f"MODEL : {model_name}")
    
    print("\nAccuracy :", accuracy)
    
    print("Precision :", precision)
    
    print("Recall :", recall)
    
    print("F1 Score :", f1)
    
    print("ROC-AUC :", roc_auc)
    
    print("\n--------------------------------------------------\n")
    
    # Classification Report
    print("Classification Report :\n")
    
    print(classification_report(y_test, y_pred))
    
    print("\n--------------------------------------------------\n")
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    disp = ConfusionMatrixDisplay(
        
        confusion_matrix=cm,
        
        display_labels=['Non-Fraud', 'Fraud']
    )
    
    disp.plot(cmap='Blues')
    
    plt.title(f'Confusion Matrix - {model_name}')
    
    plt.show()
    
    return model
```

```python
# ============================================================
# LOGISTIC REGRESSION
# ============================================================

"""
Logistic Regression is used as a baseline linear classification model.

Advantages:
- fast and interpretable
- works well on high-dimensional datasets
- useful baseline for fraud classification
"""

log_reg = LogisticRegression(
    
    random_state=42,
    
    max_iter=1000
)

log_reg_model = evaluate_model(
    
    log_reg,
    
    "Logistic Regression"
)
```

```python
# ============================================================
# RANDOM FOREST CLASSIFIER
# ============================================================

"""
Random Forest is an ensemble learning algorithm capable of:
- handling non-linear patterns
- reducing overfitting
- capturing complex fraud behavior

It is widely used in fraud detection systems.
"""

random_forest = RandomForestClassifier(
    
    n_estimators=100,
    
    random_state=42,
    
    n_jobs=-1
)

rf_model = evaluate_model(
    
    random_forest,
    
    "Random Forest"
)
```

```python
# ============================================================
# GRADIENT BOOSTING CLASSIFIER
# ============================================================

"""
Gradient Boosting builds sequential weak learners and improves
prediction performance iteratively.

Advantages:
- strong predictive capability
- handles complex fraud patterns
- effective for tabular datasets
"""

gradient_boost = GradientBoostingClassifier(
    
    random_state=42
)

gb_model = evaluate_model(
    
    gradient_boost,
    
    "Gradient Boosting"
)
```

```python
# ============================================================
# 13. HANDLING CLASS IMBALANCE USING SMOTE
# ============================================================

"""
The fraud dataset is extremely imbalanced.

Problem:
- Fraudulent transactions are very rare
- Machine learning models may become biased toward majority class
- Minority fraud cases may be ignored

Solution:
SMOTE (Synthetic Minority Oversampling Technique)

SMOTE works by:
- generating synthetic fraud samples
- improving minority class learning
- helping models better detect fraudulent behavior

Important Note:

SMOTE will ONLY be applied on training data to prevent:
- data leakage
- unrealistic performance
- test contamination

This follows industry-standard fraud detection practices.
"""
```

```python
# ============================================================
# IMPORTING SMOTE
# ============================================================

from imblearn.over_sampling import SMOTE
```

```python
# ============================================================
# CHECKING ORIGINAL CLASS DISTRIBUTION
# ============================================================

print("Original Training Target Distribution :\n")

print(y_train.value_counts())
```

```python
# ============================================================
# APPLYING SMOTE
# ============================================================

smote = SMOTE(
    
    random_state=42
)

X_train_smote, y_train_smote = smote.fit_resample(
    
    X_train,
    
    y_train
)

print("SMOTE Applied Successfully")
```

```python
# ============================================================
# CHECKING RESAMPLED DISTRIBUTION
# ============================================================

print("Resampled Training Target Distribution :\n")

print(y_train_smote.value_counts())
```

```python
# ============================================================
# VISUALIZING SMOTE DISTRIBUTION
# ============================================================

plt.figure(figsize=(8,5))

sns.countplot(
    
    x=y_train_smote
)

plt.title('Class Distribution After SMOTE')

plt.xticks(
    
    ticks=[0,1],
    
    labels=['Non-Fraud', 'Fraud']
)

plt.show()
```

```python
# ============================================================
# OBSERVATIONS ON SMOTE
# ============================================================

"""
Key Observations:

- Fraud class distribution is now balanced
- Synthetic fraud samples have been generated
- Minority class representation improved significantly

Business Importance:

- Models can now learn fraud patterns more effectively
- Fraud sensitivity is expected to improve
- Recall for fraud detection may increase

Technical Importance:

- SMOTE was applied ONLY on training data
- Test set remains untouched for realistic evaluation
- Leakage-safe workflow maintained
"""
```

```python
# ============================================================
# REINITIALIZING RESULTS STORAGE
# ============================================================

smote_results = []
```

```python
# ============================================================
# MODEL EVALUATION FUNCTION FOR SMOTE DATA
# ============================================================

def evaluate_smote_model(model, model_name):
    
    # Training Model
    model.fit(X_train_smote, y_train_smote)
    
    # Predictions
    y_pred = model.predict(X_test)
    
    # Probabilities
    y_prob = model.predict_proba(X_test)[:,1]
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    
    precision = precision_score(y_test, y_pred)
    
    recall = recall_score(y_test, y_pred)
    
    f1 = f1_score(y_test, y_pred)
    
    roc_auc = roc_auc_score(y_test, y_prob)
    
    # Train Score
    train_score = model.score(
        
        X_train_smote,
        
        y_train_smote
    )
    
    # Test Score
    test_score = model.score(
        
        X_test,
        
        y_test
    )
    
    # Storing Results
    smote_results.append({
        
        'Model' : model_name,
        
        'Accuracy' : accuracy,
        
        'Precision' : precision,
        
        'Recall' : recall,
        
        'F1 Score' : f1,
        
        'ROC-AUC' : roc_auc,
        
        'Train Score' : train_score,
        
        'Test Score' : test_score
    })
    
    # Printing Results
    print(f"MODEL : {model_name}")
    
    print("\nAccuracy :", accuracy)
    
    print("Precision :", precision)
    
    print("Recall :", recall)
    
    print("F1 Score :", f1)
    
    print("ROC-AUC :", roc_auc)
    
    print("\n--------------------------------------------------\n")
    
    print("Classification Report :\n")
    
    print(classification_report(y_test, y_pred))
    
    print("\n--------------------------------------------------\n")
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    disp = ConfusionMatrixDisplay(
        
        confusion_matrix=cm,
        
        display_labels=['Non-Fraud', 'Fraud']
    )
    
    disp.plot(cmap='Purples')
    
    plt.title(f'Confusion Matrix - {model_name}')
    
    plt.show()
    
    return model
```

```python
# ============================================================
# ROC CURVE VISUALIZATION FUNCTION
# ============================================================

def plot_roc_curve(model, model_name):
    
    # Probability Prediction
    y_prob = model.predict_proba(X_test)[:,1]
    
    # ROC Metrics
    fpr, tpr, thresholds = roc_curve(
        
        y_test,
        
        y_prob
    )
    
    roc_auc = auc(fpr, tpr)
    
    # Plot
    plt.figure(figsize=(8,6))
    
    plt.plot(
        
        fpr,
        
        tpr,
        
        label=f'{model_name} (AUC = {roc_auc:.4f})'
    )
    
    plt.plot(
        
        [0,1],
        
        [0,1],
        
        linestyle='--'
    )
    
    plt.xlabel('False Positive Rate')
    
    plt.ylabel('True Positive Rate')
    
    plt.title(f'ROC Curve - {model_name}')
    
    plt.legend()
    
    plt.grid()
    
    plt.show()
```

```python
# ============================================================
# PRECISION-RECALL CURVE FUNCTION
# ============================================================

def plot_pr_curve(model, model_name):
    
    # Probability Prediction
    y_prob = model.predict_proba(X_test)[:,1]
    
    # PR Metrics
    precision, recall, thresholds = precision_recall_curve(
        
        y_test,
        
        y_prob
    )
    
    pr_auc = auc(recall, precision)
    
    # Plot
    plt.figure(figsize=(8,6))
    
    plt.plot(
        
        recall,
        
        precision,
        
        label=f'{model_name} (AUC = {pr_auc:.4f})'
    )
    
    plt.xlabel('Recall')
    
    plt.ylabel('Precision')
    
    plt.title(f'Precision-Recall Curve - {model_name}')
    
    plt.legend()
    
    plt.grid()
    
    plt.show()
```

```python
from sklearn.metrics import average_precision_score
```

```python
# ============================================================
# MODEL EVALUATION COMPARISON FUNCTION
# ============================================================

def compare_models(
    
    model,
    
    model_name,
    
    X_train_data,
    
    y_train_data,
    
    X_test_data,
    
    y_test_data
):
    
    # Predictions
    y_pred = model.predict(X_test_data)
    
    # Probabilities
    y_prob = model.predict_proba(X_test_data)[:,1]
    
    # Metrics
    accuracy = accuracy_score(
        
        y_test_data,
        
        y_pred
    )
    
    precision = precision_score(
        
        y_test_data,
        
        y_pred
    )
    
    recall = recall_score(
        
        y_test_data,
        
        y_pred
    )
    
    f1 = f1_score(
        
        y_test_data,
        
        y_pred
    )
    
    roc_auc = roc_auc_score(
        
        y_test_data,
        
        y_prob
    )
    
    pr_auc = average_precision_score(
        
        y_test_data,
        
        y_prob
    )
    
    # Overfitting Check
    train_score = model.score(
        
        X_train_data,
        
        y_train_data
    )
    
    test_score = model.score(
        
        X_test_data,
        
        y_test_data
    )
    
    # Returning Results
    return {
        
        'Model' : model_name,
        
        'Accuracy' : accuracy,
        
        'Precision' : precision,
        
        'Recall' : recall,
        
        'F1 Score' : f1,
        
        'ROC-AUC' : roc_auc,
        
        'PR-AUC' : pr_auc,
        
        'Train Score' : train_score,
        
        'Test Score' : test_score
    }
```

```python
# ============================================================
# COMPARING ALL MODELS
# ============================================================

final_model_results = []

# Logistic Regression + SMOTE
final_model_results.append(
    
    compare_models(
        
        log_reg_smote_model,
        
        "Logistic Regression + SMOTE",
        
        X_train_smote,
        
        y_train_smote,
        
        X_test,
        
        y_test
    )
)

# Random Forest + SMOTE
final_model_results.append(
    
    compare_models(
        
        rf_smote_model,
        
        "Random Forest + SMOTE",
        
        X_train_smote,
        
        y_train_smote,
        
        X_test,
        
        y_test
    )
)

# Gradient Boosting + SMOTE
final_model_results.append(
    
    compare_models(
        
        gb_smote_model,
        
        "Gradient Boosting + SMOTE",
        
        X_train_smote,
        
        y_train_smote,
        
        X_test,
        
        y_test
    )
)

# Final Tuned Model
final_model_results.append(
    
    compare_models(
        
        best_tuned_model,
        
        "Final Optimized Gradient Boosting",
        
        X_train_sampled,
        
        y_train_sampled,
        
        X_test,
        
        y_test
    )
)
```
